```python

## lineplot()
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker='*',ms=18,mfc='yellow') 
plt.show()
# ms = markersizeabs  
# mfc = marker face color
```

```python
## lineplot()
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker='P' ,ms=18,mfc='yellow') 
plt.xlabel('Axis-X') #label the xaxis and y axis
plt.ylabel('country_population')
plt.show()
```

```python
#pieplot

country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  



plt.pie(country_population,labels=country_name)
plt.show()
```

```python

plt.pie(country_population,labels=country_name,autopct="%2.2f%%") #adding values
plt.show()
```

```python

country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  
explosion = [0.2,0,0,0,0.4]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion)
plt.show() #seperating data
```

```python
plt.figure(figsize=(6,8))   # widht and height 
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  
explosion = [0.2,0,0,0,0.4]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion,wedgeprops={'ec':'black','lw':3},shadow=True)
plt.savefig('country_pipplot')
plt.show() #underlining and shadow
```
